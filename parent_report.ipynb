```python
import json
import os
import numpy as np

class ParentReportGenerator:
    def __init__(self, brain_data, schema_path='/content/seed/parent_report_schema.json'):
        self.brain = brain_data
        self.schema = self._load_schema(schema_path)
        
    def _load_schema(self, path):
        if os.path.exists(path):
            with open(path, 'r') as f:
                return json.load(f)
        return {"learner_id": "Student_001"}

    def get_recommendation(self, weak_skill):
        """Custom suggestions based on the specific skill needing help."""
        tips = {
            "counting": "Try counting physical objects like stones or fruit at home to bridge the gap between numbers and items.",
            "addition": "Use simple kitchen activities (adding spoons or plates) to show how numbers combine.",
            "subtraction": "Play 'taking away' games with toys to help visualize subtraction.",
            "word_problem": "Read stories together and ask math questions about the characters to build logic.",
            "number_sense": "Point out numbers in the real world (on doors, buses, or prices) to build familiarity."
        }
        return tips.get(weak_skill, "Encourage 10 minutes of daily practice to build confidence.")

    def generate_markdown(self):
        learner = self.schema.get('learner_id', 'The Student')
        mastery_values = list(self.brain.skill_mastery.values())
        avg_mastery = int(np.mean(mastery_values) * 100)
        
        # Determine performance highlights
        best_skill = max(self.brain.skill_mastery, key=self.brain.skill_mastery.get)
        weak_skill = min(self.brain.skill_mastery, key=self.brain.skill_mastery.get)
        
        # Header & Overall Score
        report = f"# 📊 Mwenyura AI: Weekly Progress Report\n"
        report += f"**Learner ID:** {learner} | **Overall Progress:** {avg_mastery}% {'⭐' * (avg_mastery // 20)}\n\n"
        
        # Skill Table
        report += "| Skill Area | Progress Bar | Level |\n"
        report += "| :--- | :--- | :--- |\n"
        
        for skill, val in self.brain.skill_mastery.items():
            percentage = int(val * 100)
            # Create a 10-block progress bar
            bar_size = max(1, int(val * 10))
            bar = "🟩" * bar_size + "⬜" * (10 - bar_size)
            
            skill_name = skill.replace('_', ' ').title()
            report += f"| **{skill_name}** | {bar} | {percentage}% |\n"
        
        # Recommendations & Insights
        report += f"\n---\n"
        report += f"### 💡 Personalized Insights\n"
        report += f"🌟 **Top Strength:** Doing amazing in **{best_skill.title()}**! Your child shows great speed and accuracy here.\n\n"
        report += f"⚠️ **Focus Area:** We noticed a need for more support in **{weak_skill.replace('_', ' ')}**.\n\n"
        report += f"📖 **Recommendation:** {self.get_recommendation(weak_skill)}\n\n"
        
        report += f"---\n*Generated by Mwenyura AI Tutor Platform*"
        return report

# --- EXAMPLE USAGE (Integration with your Brain class) ---
# If you want to test this stand-alone:
if __name__ == "__main__":
    # Mock data to simulate the Tutor Brain
    class MockBrain:
        def __init__(self):
            self.skill_mastery = {
                "counting": 0.85,
                "addition": 0.40,
                "subtraction": 0.30,
                "word_problem": 0.20,
                "number_sense": 0.60
            }

    mock_brain = MockBrain()
    generator = ParentReportGenerator(mock_brain)
    print(generator.generate_markdown())
```
