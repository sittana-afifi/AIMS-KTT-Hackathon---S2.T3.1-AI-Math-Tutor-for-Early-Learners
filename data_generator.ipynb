```python
import shutil
import os

# 1. Get the path from kagglehub
import kagglehub
downloaded_path = kagglehub.dataset_download("nirmalsankalana/plant-diseases-training-dataset")

# 2. Define where you want it (your visible project folder)
destination = "/content/ktt"

# 3. Move it so you can see it in the sidebar
if not os.path.exists(destination):
    shutil.copytree(downloaded_path, destination)
    print(f"✅ Data moved to: {destination}")
else:
    print("✅ Data already exists in /content")
```

```python
import os
import shutil
# Filtter the data
# 1. Path setup
source_base = "/content/ktt/data"
# Check if there's a 'train' or 'data' subfolder inside
if "train" in os.listdir(source_base):
    source_base = os.path.join(source_base, "train")
elif "data" in os.listdir(source_base):
    source_base = os.path.join(source_base, "data")

target_base = "/content/ktt/data/mini_plant_set"
os.makedirs(target_base, exist_ok=True)

# 2. Define what we are looking for (Keywords)
# We will search for these strings in your actual folder names
search_targets = {
    "maize_rust": ["Corn", "rust"],
    "maize_blight": ["Corn", "Blight"],
    "maize_healthy": ["Corn", "healthy"],
    "cassava_mosaic": ["Cassava", "Mosaic"],
    "bean_spot": ["Bean", "Spot"]
}

print(f"--- Searching in: {source_base} ---")
actual_folders = os.listdir(source_base)

for clean_label, keywords in search_targets.items():
    # Find the folder that contains ALL keywords for this class
    match = None
    for folder in actual_folders:
        if all(k.lower() in folder.lower() for k in keywords):
            match = folder
            break

    if match:
        src_path = os.path.join(source_base, match)
        dest_path = os.path.join(target_base, clean_label)
        os.makedirs(dest_path, exist_ok=True)

        # Get images and limit to 300
        images = [f for f in os.listdir(src_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        selected = images[:300]

        for img in selected:
            shutil.copy(os.path.join(src_path, img), os.path.join(dest_path, img))

        print(f"✅ Matched '{match}' -> Saved as '{clean_label}' ({len(selected)} images)")
    else:
        print(f"❌ NOT FOUND: Keywords {keywords}")

print(f"\n🎉 Clean dataset ready at: {target_base}")
```
